```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data-science-youtube-channel-comparison.csv')
```

### **Here we can see that in total we have 51 data science channels and for all the channels we have 6 features representing analytics of that channels.**

## The 6 features are : 

1.) **channel name** : Name of the channel(str)

2.) **views** : The total number of views on that channel upto '2022-01-28' in(int)

3.) **subs** : Number of Subscribers(int)

4.) **last 30 days views** : Aggregate of views on all his videos for the last 30 days(int)

5.) **num videos** : Total number of videos in his/her channel(int)

6.) **how old** : When was the channel created(str). This data contains the date on which he/she uploaded his/her first video.

```python
df['views'] = df['views'] * 1000000
df['subs'] = df['subs'] * 1000
df['last 30 days views'] = df['last 30 days views']*1000
```

```python
df.head()
```

```python
df['how old'] = pd.to_datetime(df['how old'])
basedate = pd.Timestamp('2022-01-28')
df['channel_age'] = (basedate-df['how old']).dt.days
```

```python
df.head()
```

```python
df.drop('how old',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'channel name':'channel_name','last 30 days views':'last_30_days_views','num videos':'total_no_of_videos'},inplace=True)
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

***Here onwards I am going to compare CampusX with other 50 channels using 7 Metrics***

### View Metirc

```python
# top 10 YouTube channels according to views.
df.sort_values('views',ascending=False)[['channel_name','views']].head(10)
```

### Views/Sub count Metric
